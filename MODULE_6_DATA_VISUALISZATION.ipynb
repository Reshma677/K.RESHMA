5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data with correlated features
num_samples = 100
feature1 = np.random.rand(num_samples)
feature2 = feature1 + np.random.randn(num_samples) * 0.2  # Correlated with feature1
feature3 = feature1 - np.random.randn(num_samples) * 0.3  # Negatively correlated with feature1
feature4 = np.random.rand(num_samples)  # Uncorrelated with others

# Create a Pandas DataFrame
data = pd.DataFrame({'Feature1': feature1, 'Feature2': feature2, 'Feature3': feature3, 'Feature4': feature4})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Define the x and y ranges
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid from x and y
x, y = np.meshgrid(x, y)
# Calculate z values using the function
z = np.sin(np.sqrt(x**2 + y**2))

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
# Update layout for better visualization
fig.update_layout(title='3D Surface Plot of z = sin(√x² + y²)',
                  autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))
# Display the plot
fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python


import numpy as np
from bokeh.plotting import figure, show

# Create x and y data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)

# Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line glyph to the plot
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

2..Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Generate random data
num_points = 50
x = np.random.rand(num_points) * 10  # x values between 0 and 10
y = np.random.rand(num_points) * 10  # y values between 0 and 10
sizes = np.random.randint(5, 20, size=num_points)  # Random sizes between 5 and 20
colors = np.random.choice(['red', 'blue', 'green', 'yellow'], size=num_points)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a Bokeh plot
p = figure(title="Scatter Plot with Random Data", x_axis_label="x", y_axis_label="y")

# Add a scatter glyph to the plot
p.scatter(x='x', y='y', size='sizes', color='colors', source=source)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

         fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
         counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Define the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh plot
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label="Fruit", y_axis_label="Count")

# Add a bar glyph to the plot
p.vbar(x='fruits', top='counts', width=0.9, source=source)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Assuming you have data_hist and hist, edges from your previous code
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)


# Create a ColumnDataSource
source = ColumnDataSource(data=dict(top=hist, bottom=[0]*len(hist), left=edges[:-1], right=edges[1:]))

# Create a Bokeh plot
p = figure(title="Histogram of Data", x_axis_label="Values", y_axis_label="Frequency")

# Add a quad glyph for the histogram bars
p.quad(top='top', bottom='bottom', left='left', right='right',
       fill_color="skyblue", line_color="white", source=source)
# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python


import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.palettes import Viridis256  # Or any other palette you prefer

# Assuming you have data_heatmap, x, y, xx, yy from your previous code
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=xx.flatten(), y=yy.flatten(), values=data_heatmap.flatten()))

# Create a color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create a Bokeh plot
p = figure(title="Heatmap", x_axis_label="x", y_axis_label="y",
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           tools="hover", tooltips=[("x", "@x"), ("y", "@y"), ("value", "@values")])  # Add hover tooltips

# Add a rect glyph for the heatmap cells
p.rect(x='x', y='y', width=0.1, height=0.1,
       fill_color={'field': 'values', 'transform': color_mapper},
       line_color="white", source=source)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12)
p.add_layout(color_bar, 'right')  # Place color bar on the right

# Show the plot
show(p)
```
